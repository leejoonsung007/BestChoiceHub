```python
for school_name in school_names:
    if 'Baile Atha Cliath' in school_name:
        school_name = school_name.replace('Baile Atha Cliath','Dublin')
        result = gmaps.geocode(school_name)
        if len(result) != 0:
            print((result[0])['place_id'])
            place_id.append ((result[0])['place_id'])
        else:   
            place_id.append('No Data')
    else:
        result = gmaps.geocode(school_name)
        if len(result) != 0:
            print((result[0])['place_id'])
            place_id.append ((result[0])['place_id'])
        else:   
            place_id.append('No Data')
```
