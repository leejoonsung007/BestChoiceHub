```python
# use place_id to get the details of the place and get the photo reference
place_id_list = []
place_name_list = []
author_name_list = []
author_avatar_list = []
rating_list = []
review_list = []
overall_rating_list = []
time_list = []
relative_time_description_list = []

for id in place_id:
    response = gmaps.place(place_id=id)
    if 'reviews' in response['result']:
        review_responses = response['result']['reviews']
    else:
        pass
    name = response['result']['name']
    
    if 'rating' in response['result']:
        overall_rate = response['result']['rating']
    else:
        overall_rate = 'No Data'
    
    print(review_responses)
    for review_response in review_responses:
        place_id_list.append(id)
        place_name_list.append(name)
        
        author_name = review_response['author_name']
        author_name_list.append(author_name)
        
        if 'profile_photo_url' in review_response:
            author_avatar = review_response['profile_photo_url']
            author_avatar_list.append(author_avatar)
        else:
            author_avatar = review_response['profile_photo_url']
            author_avatar_list.append('No Avatar')
            
        
        
        rating = review_response['rating'] 
        rating_list.append(rating)
        
        overall_rating_list.append(overall_rate)
        
        review = review_response['text']
        if review == '':
            review_list.append('The user rated the school but without any comment')
        else:
            review_list.append(review)
            
        relative_time_description = review_response['relative_time_description']
        relative_time_description_list.append(relative_time_description)
            
        time = review_response['time']
        format_time = pd.to_datetime(time, unit='s')
        time_list.append(format_time)
        
#     if review is None:
#         review = 'No data'

    
#     name = (response['result'])['name']
#     names.append(name)
    
#     try:
#         photos = (response['result'])['photos']
#         for photo in photos:
#             name_photoref[name].append(photo['photo_reference'])
#     except:
#         pass
    
#     values = name_photoref[name]
    
#     values_list.append(values)
```

```python
df1 = pd.DataFrame(place_id_list)
df2 = pd.DataFrame(place_name_list)
df3 = pd.DataFrame(author_name_list)
df4 = pd.DataFrame(author_avatar_list)
df5 = pd.DataFrame(rating_list)
df6 = pd.DataFrame(review_list)
df7 = pd.DataFrame(overall_rating_list)
df8 = pd.DataFrame(time_list)
df9 = pd.DataFrame(relative_time_description_list)
```
