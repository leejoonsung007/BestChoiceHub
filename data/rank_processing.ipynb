```python
place_id = []
```

```python
for school_name in school_names:
    result = gmaps.geocode(school_name)
    print((result[0])['place_id'])
    place_id.append ((result[0])['place_id']) 
```

```python
len(place_id)
```

```python
names = []
```

```python
values_list = []
```

```python
# use place_id to get the details of the place and get the photo reference
name_photoref = defaultdict(list)
for id in place_id:
    response = gmaps.place(place_id=id)
    name = (response['result'])['name']
    names.append(name)   
    try:
        photos = (response['result'])['photos']
        for photo in photos:
            name_photoref[name].append(photo['photo_reference'])
    except:
        pass
    
    values = name_photoref[name]
    print(values)
    
    values_list.append(values)
```

```python
df = pd.DataFrame(values_list)
```
